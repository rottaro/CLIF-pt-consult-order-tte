```python
import pandas as pd
block_final = pd.read_parquet('block_compiled.parquet')

#Add in a dichotomized outcomes variables
block_final['pt_post48_Yellow'] = block_final['delayed_yellow_PT'].notna() & (block_final['delayed_yellow_PT'] <= 48)
block_final['pt_pre24_IMV'] = block_final['Time_last_PT'].notna() & (block_final['Time_last_PT'] >= -24)
block_final['pt_pre24_or_post48_Yellow'] = block_final['pt_pre24_IMV'] | block_final['pt_post48_Yellow']

block_final['duplicate_patient'] = block_final.duplicated(subset=['patient_id'], keep=False)

dup_df = block_final[block_final['duplicate_patient']]

print(dup_df.shape)

pivot_df = dup_df.pivot_table(index="patient_id",
                values="pt_pre24_or_post48_Yellow",
                aggfunc={"pt_pre24_or_post48_Yellow": ["count","sum","mean"]})


print(pivot_df.shape)
print(pivot_df['count'].describe())
print(pivot_df['mean'].describe())
pivot_df['exp'] = (pivot_df['mean'] > 0 ) & (pivot_df['mean'] < 1 )
print(pivot_df['exp'].sum())
```
