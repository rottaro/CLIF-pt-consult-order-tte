```python
#Graph by year
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet('block_with_data.parquet')

pivot_df = df.pivot_table(index="admission_year",
                values="pt_pre24_or_post48_Yellow",
                aggfunc="sum")

pivot_df = df.pivot_table(index="admission_year",
                values="pt_pre24_or_post48_Yellow",
                aggfunc={"pt_pre24_or_post48_Yellow": ["count","sum","mean"]})
pivot_df.rename(columns={"count":"N", "sum":"Early_PT_N", "mean":"Early_PT"}, inplace=True)

pivot_df = pivot_df.reset_index()
pivot_df["admission_year"] = pivot_df["admission_year"].astype(str)

pivot_df.head(20)
#pivot_df.to_csv('Year_Trend.csv')
```

```python
# Create the bar graph
plt.bar(pivot_df["admission_year"], pivot_df["Early_PT"] , color='skyblue')

# Add labels and title
plt.xlabel('Admission Year')
plt.xticks(rotation='vertical')
plt.ylabel('Prevelance of Early PT Orderr')

# Display the graph
plt.show()

plt.savefig('Year_Trend.png') 
```
